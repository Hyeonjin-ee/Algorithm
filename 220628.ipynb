#### 백준 2525번

```python
hour, minute = map(int, input().split())
required_minutes = int(input())

print(hour, minute)
print(required_minutes)

hour = (hour + (minute + required_minutes)//60)%24
minute = (minute + required_minutes)%60
print(hour, minute)
```

#### 백준 2480번

```python
x, y, z = map(int, input().split())

if (x == y and x == z) :
    print(10000 + x * 1000)
elif(x == y or x == z) :
    print(1000 + x * 100)
elif(y == z) :
    print(1000+y*100)
else :
    max = max(x,y,z)
    print(max * 100)
```
